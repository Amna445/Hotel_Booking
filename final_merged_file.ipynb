# A thorough analysis and modelling of cancellations of hotel reservations

In the hospitality sector, demand management choices are greatly influenced by booking cancellations.
More than 140 million reservations are made online each year, with many of those for hotels coming from popular travel sites like Booking.com, Expedia.com, Hotels.com, etc.
The average cancellation rate for bookings worldwide has risen to about 40%, according to D-Edge Hospitality Solutions' analysis of data from the previous five years. This trend has a particularly negative effect on hotel revenue and distribution management methods.
Hotels adopt strict cancellation rules, inventory management, and overbooking methods, which can potentially have a detrimental effect on revenue and reputation, to address the issues brought on by booking cancellations.

There is not much that can be done after a reservation has been canceled, which is uncomfortable for many hotels and hotel technology businesses. Therefore, foreseeing potential cancellations of reservations and preventing them would generate extra revenue for both hotels and hotel technology businesses.

Fortunately, we can estimate which customers are most likely to cancel a reservation using Python and machine learning, which could lead to more accurate projections and less ambiguity in business decisions. Therefore, our business questions are:

   ### *1. Predicting whether the hotel bookings made by guest will be canceled or not?*
   ### *2. Identify profitable customers by their market segmentations?*
   ### *3. Predicting the price of a hotel booking made by a customer.*

## 2. Data Pre-processing

### 2.1 Missing values

```python
percentNA = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
percentNA
```

## Key Insights:
- `company` column needs to be dropped due to large % of NA values.
- `agent` column, that has missing values implies that the booking is without any agent thus all NA values should be replaced by 0
- `country` columns that has missing rows, needs to be dropped because we cant fill-up the origin country
- `children` columns that has missing rows, needs to be dropped because we cant be sure about that count.

```python
# dropping company column
df = df.drop(['company'], axis=1)
# filling the agent missing values with 0
df['agent'].fillna(0, inplace=True)
# droping the country missing rows
df.dropna(subset=['country'], inplace=True)
# dropping the children missing rows
df.dropna(subset=['children'], inplace=True)

# checking the percentage of missing values in each column again
percentNA = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
print(percentNA)

print("data has {} rows and {} columns".format(df.shape[0], df.shape[1]))
```

### 2.2 Inconsistent information

```python
# lets first rename some columns for further ease.
df.rename(columns={'arrival_date_year': 'year',
                'arrival_date_month': 'month',
                'arrival_date_week_number': 'week_number',
                'arrival_date_day_of_month': 'day',
                }, inplace=True)
```

- ***Feature Engineering***

```python
## making some new columns

#Create datetime value

df['month'] = df['month'].map({'January':1, 'February': 2, 'March':3, 'April':4, 'May':5, 'June':6, 'July':7,
                            'August':8, 'September':9, 'October':10, 'November':11, 'December':12})

#Create a total guests column
df['total_guests'] = df['adults'] + df['children'] + df['babies']

#Create a total nights stay column
df['total_nights_stay'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']

#creating a cancel rate column by dividing the cancels by the total bookings that are not canceled
df['cancel_rate'] = df['previous_cancellations'] / \
    (df['previous_cancellations'] + df['previous_bookings_not_canceled'])

df['cancel_rate'].fillna(0, inplace=True)

# creating total previous bookings
df['total_previous_bookings'] = df['previous_cancellations'] + df['previous_bookings_not_canceled']
```

2.2.1:

```python
# checking for rows that have adults = 0 and children = 0 and babies = 0 at the same time
df[(df['adults'] == 0) & (df['children'] == 0) & (df['babies'] == 0)].shape
```

## Key insight:
- These 170 observations are not making any sense i.e., there is no adult, children or baby in the hotel listing or booking. Therefore, better to remove these rows.

```python
# removing those rows
df = df[(df['adults'] != 0) | (df['children'] != 0) | (df['babies'] != 0)]
print("data has {} rows and {} columns".format(df.shape[0], df.shape[1]))
```

2.2.2:

```python
# removing the rows with adr is negative
df = df[df['adr'] > 0]
```

## Key insight:
- The average daily rate should be +ve and 0 in a case if a voucher has redeemed. Therefore, dropping negative `adr` values.

2.2.3:

```python
#Where the customer is listed as a repeated guest, but has no previous bookings
#they will be changed to 0.
df['is_repeated_guest'] = np.where(df['total_previous_bookings'] == 0\
                            , 0, df['is_repeated_guest'])
```

```python
#Where the customer is listed as a first time guest, but has previous bookings
#they will be changed to 1.
df['is_repeated_guest'] = np.where(df['total_previous_bookings'] > 0\
                            , 1, df['is_repeated_guest'])
```

## Key insight:
- A lot of values were observed with wrong consistent information wrt previous bookings, repeated guests, cancelation rates. Therefore, guests that are listed as a repeated guests with no previous bookings changed to 0 and vice versa.

```python
# multicollinearlity check
# function to get the correlation of each feature with one another
def correlation(dataset, threshold):
    col_corr = set() # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i] # getting the name of column
                col_corr.add(colname)
    return col_corr

# print the correlated features
correlation(df_transformed, 0.5)
```

In this instance, K-Means is used to analyse market segment clusters for a hotel in Portugal.

Given lead time (the period of time from when the customer makes their booking to when they actually stay at the hotel), along with ADR (average daily rate per customer), the k-means clustering algorithm is used to visually identify which market segments are most profitable for the hotel.

A customer with a high ADR and a low lead time is ideal, as it means that 1) the customer is paying a high daily rate which means a greater profit margin for the hotel, while a low lead time means that the customer pays for their booking quicker — which increases cash flow for the hotel in question.



#### The interval (or continuous random variables) are of lead time and ADR are defined as below.
#### Variables with a categorical component are defined using ‘’’cat.codes’’’, in this case market segment.

```python
from sklearn.model_selection import train_test_split

# target feature as "y" and the other features as "x"
x = df1.drop('adr', axis=1)
y = df1['adr']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state= 95)
```

```python
#Import the scikit-learn functions and classes necessary to perform cross-validation
from sklearn.pipeline import make_pipeline
from sklearn import preprocessing
from sklearn.model_selection import GridSearchCV

# Import the scikit-learn class used to train an Elastic-Net model
from sklearn.linear_model import ElasticNet

# Create a pipeline specifying all of the operations to perform when training the model
# In this case, the pipepline consists of z-score standardization and fitting of an Elastic-Net model
pipeline_en = make_pipeline(preprocessing.StandardScaler(), ElasticNet(fit_intercept = True))

# Specify the hyperparameters and their corresponding values that are to be used in GridSearch
hyperparameters_en = { 'elasticnet__alpha' : [0.1, 0.3, 0.5, 0.7, 0.9, 1],
                  'elasticnet__l1_ratio': [0.1, 0.3, 0.5, 0.7, 0.9, 1]}

# Initialize the GridSearch cross-validation object, specifying 10 folds for 10-fold cross-validation and
# "r2" and "neg_mean_squared_error" as the evaluation metrics for cross-validation scoring
elastic_net = GridSearchCV(pipeline_en, hyperparameters_en, cv = 10, scoring = ['r2', 'neg_mean_squared_error'],
                            refit = 'r2', verbose = 0, n_jobs = -1)

# Train and cross-validate the random forest regression model and ignore the function output
_ = elastic_net.fit(X_train, y_train)
```

```python
print("Elastic Net Regression Accuracy Score (Train) :", elastic_net.score(X_train,y_train))
print("Elastic Net Regression Accuracy Score (Test) :",elastic_net.score(X_test,y_test))
```

```python
# Import the scikit-learn class used to train a random forest regression model
from sklearn.ensemble import RandomForestRegressor

# Create a pipeline specifying all of the operations to perform when training the model
# In this case, the pipepline consists of z-score standardization and fitting of a random forest regressor
pipeline_rf = make_pipeline(preprocessing.StandardScaler(),
                            RandomForestRegressor(n_estimators=100, random_state = 42))

# Specify the hyperparameters and their corresponding values that are to be used in GridSearch
hyperparameters_rf = {'randomforestregressor__max_features' : ['auto', 'sqrt', 'log2'],
                      'randomforestregressor__max_depth': [None, 5]}

# Initialize the GridSearch cross-validation object, specifying 10 folds for 10-fold cross-validation and
# "r2" and "neg_mean_squared_error" as the evaluation metrics for cross-validation scoring
random_forest = GridSearchCV(pipeline_rf, hyperparameters_rf, cv = 10, scoring = ['r2', 'neg_mean_squared_error'],
                            refit = 'r2', verbose = 0, n_jobs = -1)

# Train and cross-validate the random forest regression model and ignore the function output
_ = random_forest.fit(X_train, y_train)
```

```python
print("Random Forest Regressor Accuracy Score (Train) :", random_forest.score(X_train,y_train))
print("Random Forest Regressor Accuracy Score (Test) :",random_forest.score(X_test,y_test))
```

```python
# Import the scikit-learn functions used to calculate the R^2 and NMSE on the test set
from sklearn.metrics import r2_score, mean_squared_error

# Use the best Elastic-Net model to make predictions on the test set
y_test_pred_en = elastic_net.predict(X_test)

# Display the R^2 and NMSE on the train and test sets for the Elastic-Net model
print('Elastic-Net R^2 (train):', round(elastic_net.cv_results_['mean_test_r2'][elastic_net.best_index_], 3))
print('Elastic-Net R^2 (test):', round(r2_score(y_test, y_test_pred_en), 3), '\n')
print('Elastic-Net NMSE (train):', round(elastic_net.cv_results_['mean_test_neg_mean_squared_error'][elastic_net.best_index_], 3))
print('Elastic-Net NMSE (test):', round(mean_squared_error(y_test, y_test_pred_en) * -1, 3), '\n')

# Use the best random forest model to make predictions on the test set
y_test_pred_rf = random_forest.predict(X_test)

# Display the R^2 and NMSE on the train and test sets for the random forest model
print('Random forest R^2 (train):', round(random_forest.cv_results_['mean_test_r2'][random_forest.best_index_], 3))
print('Random forest R^2 (test):', round(r2_score(y_test, y_test_pred_rf), 3), '\n')
print('Random forest NMSE (train):', round(random_forest.cv_results_['mean_test_neg_mean_squared_error'][random_forest.best_index_], 3))
print('Random forest NMSE (test):', round(mean_squared_error(y_test, y_test_pred_rf) * -1, 3))
```
